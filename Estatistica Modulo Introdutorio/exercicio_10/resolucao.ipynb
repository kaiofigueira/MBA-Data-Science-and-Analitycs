# Enunciado

(Fonte: Fávero e Belfiore, 2024, Cap. 7) Um fabricante de iogurtes 
desnatados afirma que a quantidade de calorias em cada pote é 60 cal. Para 
verificar se essa informação procede, uma amostra aleatória com 36 potes é 
coletada, observando-se que a quantidade média de calorias é de 65 cal com 
desvio-padrão 3,5. Aplique o teste adequado e verifique se a afirmação do 
fabricante é verdadeira, considerando o nível de significância de 5%. 

# Imports

```python
from scipy.stats import t
import math
```

# Leitura dos dados

```python
alpha = 0.05
media_amostral = 65
desvio_padrao_amostral = 3.5
n = 36
mu_0 = 60
```

# Resolução

cada pote tem mesmo 60 cal em media?

H0: media = 60

H1: media != 60

```python
t_calculado = (media_amostral - mu_0) / (desvio_padrao_amostral / math.sqrt(n))
print(f"t calculado: {t_calculado}")

gl = n - 1

t_critico_lower = t.ppf(alpha / 2, gl)
t_critico_upper = t.ppf(1 - alpha / 2, gl)
print(f"Valor das duas caudas: {t_critico_lower:.4f}, {t_critico_upper:.4f}")

p_valor = 2 * (1 - t.cdf(abs(t_calculado), gl)) # teste bi-caudal
print(f"P_valor (Teste t): {p_valor:.20f}")

decisao_t_critico = "Rejeita H0" if t_calculado < t_critico_lower or t_calculado > t_critico_upper else "Não rejeita H0"
print(f"Decisão (Teste t, comparação com t critico): {decisao_t_critico}")

decisao_p_valor = "Rejeita H0" if p_valor / 2 < alpha else "Não rejeita H0"
print(f"Decisão (Teste t, comparação com p_valor): {decisao_p_valor}")
```

Conclusão: como o valor calculado para a estatística t está na região crítica, rejeita-se a hipótese nula e conclui-se que as médias são estatisticamente diferentes.
