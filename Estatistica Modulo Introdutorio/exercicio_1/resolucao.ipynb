# Enunciado
Na análise de concessão de empréstimos, uma variável potencialmente 
importante é a renda da pessoa. O gerente de um banco coleta uma base de 
dados de seus correntistas e extrai a variável “renda mensal (R$)” para 50 
pessoas. Embora se trate de uma variável quantitativa, deseja realizar uma 
análise por meio de tabela de frequências. Neste sentido, pede-se:  
a) Classifique os correntistas em faixas de renda, sendo: 0-2.000; 2.001-4.000; 
4.001-6.000; 6.001-8.000; 8.001-10.000 e 10.001-12.000.   
b) Em seguida, elabore a tabela de frequências para as faixas de renda acima.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Leitura dos dados para resolução do exercício

```python
df = pd.read_csv("dados.csv", sep=';')
df
```

# Resolução da letra a

```python
df['Renda (R$)'].dtype
```

Como renda é do tipo Object, irei converter para float para conseguir criar as faixas de renda

```python
# Primeiro, troca a vírgula por nada (remove vírgula de milhar), depois transforma em float
df['Renda (R$)'] = df['Renda (R$)'].str.replace(',', '', regex=False).astype(float)
```

Criando as faixas de renda

```python
# Definindo faixas
bins = [0, 2000, 4000, 6000, 8000, 10000, 12000]
labels = ['0-2000', '2001-4000', '4001-6000', '6001-8000', '8001-10000', '10001-12000']

# Classificando nas faixas
df['Faixa Renda'] = pd.cut(df['Renda (R$)'], bins=bins, labels=labels, right=True)
df
```

# Resolução da letra b

Criando a tabela de frenquência

```python
frequencia_absoluta = df['Faixa Renda'].value_counts().sort_index()
frequencia_absoluta
```

```python
frequencia_relativa = df['Faixa Renda'].value_counts(normalize=True).sort_index() * 100
frequencia_relativa
```

```python
tabela_freq = pd.DataFrame({
    'Frequência Absoluta': frequencia_absoluta,
    'Frequência Relativa (%)': frequencia_relativa.round(1)
})
tabela_freq
```

```python
# Frequência acumulada
tabela_freq['Frequência Acumulada'] = tabela_freq['Frequência Absoluta'].cumsum()
tabela_freq['Frequência Relativa Acumulada (%)'] = tabela_freq['Frequência Relativa (%)'].cumsum().round(1)
tabela_freq
```

```python
#Formatando a tabela
tabela_freq = tabela_freq.reset_index()
tabela_freq
```

# Visualização gráfica

```python
plt.figure(figsize=(16, 6))

# Gráfico 1: Frequência Absoluta
plt.subplot(1, 2, 1)
plt.bar(tabela_freq['Faixa Renda'], tabela_freq['Frequência Absoluta'], color='skyblue', edgecolor='black')
plt.title('Distribuição de Renda por Faixa (Frequência Absoluta)')
plt.xlabel('Faixa de Renda (R$)')
plt.ylabel('Frequência Absoluta')
plt.xticks(rotation=45)

# Gráfico 2: Frequência Relativa
plt.subplot(1, 2, 2)
plt.bar(tabela_freq['Faixa Renda'], tabela_freq['Frequência Relativa (%)'], color='lightcoral', edgecolor='black')
plt.title('Distribuição de Renda por Faixa (Frequência Relativa)')
plt.xlabel('Faixa de Renda (R$)')
plt.ylabel('Frequência Relativa (%)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
